```python
import cv2
import numpy as np

def create_image(width, height, color=(255, 255, 255)):
    """Create a blank image with the specified color."""
    image = np.zeros((height, width, 3), np.uint8)
    image[:] = color
    return image

def add_text(image, text, position, font=cv2.FONT_HERSHEY_SIMPLEX, font_scale=1, color=(0, 0, 0), thickness=2):
    """Add text to the image."""
    cv2.putText(image, text, position, font, font_scale, color, thickness)

def add_rectangle(image, top_left, bottom_right, color=(0, 0, 0), thickness=2):
    """Add a rectangle to the image."""
    cv2.rectangle(image, top_left, bottom_right, color, thickness)

def add_circle(image, center, radius, color=(0, 0, 0), thickness=2):
    """Add a circle to the image."""
    cv2.circle(image, center, radius, color, thickness)

def save_image(image, path):
    """Save the image to the specified path."""
    cv2.imwrite(path, image)

if __name__ == "__main__":
    # Create a blank image
    width, height = 800, 600
    background_color = (255, 255, 255)  # White background
    image = create_image(width, height, background_color)

    # Add some text
    add_text(image, "Hello, World!", position=(50, 100), font_scale=2, color=(0, 0, 255))

    # Add a rectangle
    add_rectangle(image, top_left=(50, 150), bottom_right=(300, 300), color=(0, 255, 0), thickness=3)

    # Add a circle
    add_circle(image, center=(400, 400), radius=50, color=(255, 0, 0), thickness=5)

    # Save the image
    output_path = "generated_image.jpg"
    save_image(image, output_path)

    print(f"Image saved to {output_path}")
```
